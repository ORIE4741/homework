```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
np.random.seed(4741)
```

```python
def f(k):
    return math.comb(6, k) * (0.5**6)

def sigma_sum(start, end, expression):
    return sum(expression(i) for i in range(start, end+1))
```

```python
def accurate(x):
    part = sigma_sum(math.ceil(3+6*x), 6, f) + sigma_sum(0, math.floor(3-6*x), f)
    return (2 * part) - part**2
```

```python
accurate(0.8)
```

```python
def h(x):
    part = 2*math.exp(-2*6*(x**2))
    return 2 * part
```

```python
h(0.3)
```

```python
hfunc = np.vectorize(h)
afunc = np.vectorize(accurate)
```

```python
from matplotlib.pyplot import figure

figure(figsize=(15, 12), dpi = 80)

x = np.linspace(0, 1, 1000)

plt.plot(x, hfunc(x), color='red', label = "Hoeffding Bound")
plt.plot(x, afunc(x), color='blue', label = "True Probability")

plt.xlabel('epsilon')
plt.ylabel('Probability')
plt.legend(loc="upper right")

plt.show()

```
